#### 4. Import the users table 

```python
# creating the query inside and using ENGINE
df_users = pd.read_sql_query('SELECT * FROM users', engine)
df_users.head(10)
```

#### 5. Rename Id column to userId

```python
#changing the name of the column wirh RENAME and replacing it with INPLACE
df_users.rename(columns={"Id": "userId"}, inplace=True)
df_users.head(10)
```

#### 6. Import the posts table. 

```python
#the same but in anither table
df_posts = pd.read_sql_query('SELECT * FROM posts', engine)
df_posts.head(10)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
df_posts.rename(columns={"Id": "postID", "OwnerUserId":"userID"}, inplace=True)
```

```python
list(df_posts.columns) #I can see how the name columns changed
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
#new_df = pd.read_sql_query('SELECT users.Id, Reputation, Views, UpVotes, DownVotes, posts.Id, Score, users.Id, ViewCount, CommentCount FROM users JOIN posts on users.Id = posts.OwnerUserId', engine)
#it's not necessary to use a query, just select the column of the dataframe and assing it to a new one
new_df_users = df_users[['userId', 'Reputation','Views','UpVotes','DownVotes']]
new_df_posts = df_posts[['postID', 'Score','userID','ViewCount','CommentCount']]
```

```python
new_df_users 
```

```python
new_df_posts
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#using MERGE function, with UserId as 'on'
new_df = new_df_users.merge(new_df_posts, how = 'inner', left_on='userId', right_on='userID')
new_df.head
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
missing = new_df.isnull().sum()
missing
#there are 48396 missing values in VIEWCOUNT
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#I'd assign zero to missing values, supposing that missing values could mean no views 
new_df['ViewCount'] = new_df['ViewCount'].fillna(0)
new_df
```
